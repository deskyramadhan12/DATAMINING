```python
#Mengecek jumlah data yang duplikat
data.duplicated().sum()
```

```python
data.drop(['id'],axis=1,inplace=True)
```

```python
objek = data.select_dtypes(include = "object").columns
print (objek)
```

```python
#Label Encoding for object to numeric conversion
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in objek:
    data['diagnosis'] = le.fit_transform(data['diagnosis'].astype(str))

print (data.info())
```

```python
x = data.drop(columns=['diagnosis'])
y = data['diagnosis']

len(y)
```

```python
corr = x.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(corr, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 15},
           cmap= 'BrBG')
plt.title('Correlation between features');
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=3)
```

```python
#import Scaler 
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
model_svc = SVC(random_state=0)
model_svc.fit(x_train,y_train)
```

```python
y_pred = model_svc.predict(x_test)
y_pred
```

```python
matrix = confusion_matrix(y_test, y_pred)
matrix
```

```python
sns.heatmap(matrix, annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print("accuracy : ", model_svc.score(x_test,y_test))
```
